```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

```python
df = pd.DataFrame({
    'SVTYPE': ["DEL", "DUP", "INS"],
    'SG10K-SV r1.3': [85195, 32464, 35670],
    'Gnomad-SV': [19274, 7479, 12584]
})
```

```python
n = 3


sg10k_color = sns.color_palette("pastel", n )
sg10k_color 
gnomad_color = sns.color_palette("bright", n )
gnomad_color 
```

```python
# https://www.webucator.com/article/python-color-constants-module/
palette_1 = sns.color_palette(["#98F5FF","#CAFF70","#FFC125"])

palette_2 = sns.color_palette(["#00BFFF","#3D9140","#ED9121","#f1c0e8","#cfbaf0","#a3c4f3","#90dbf4","#8eecf5","#98f5e1","#b9fbc0"])
palette_2

legend_elements = [Line2D([0], [0], color=palette_1[0], lw=10, label='DEL SG10K-SV (85,195)'),
                   Line2D([0], [0], color=palette_2[0], lw=10, label='DEL Overlap Gnomad-SV (19,274)'),
                   Line2D([0], [0], color=palette_1[1], lw=10, label='DUP SG10K-SV (32,464)'),
                   Line2D([0], [0], color=palette_2[1], lw=10, label='DUP Overlap Gnomad-SV (7,479)'),
                   Line2D([0], [0], color=palette_1[2], lw=10, label='INS SG10K-SV(35,670)'),
                   Line2D([0], [0], color=palette_2[2], lw=10, label='INS Overlap Gnomad-SV (12,584)')
]
```

```python
fig, ax = plt.subplots()

# plotting columns
ax = sns.barplot(x=df["SVTYPE"], y=df["SG10K-SV r1.3"], color='b' , palette = palette_1  , label="SG10K-SV r1.3" )
ax = sns.barplot(x=df["SVTYPE"], y=df["Gnomad-SV"], color='r' , palette = palette_2 , label= "Gnomad-SV")
 
# renaming the axes
ax.set(xlabel="SVTYPE", ylabel="Count")

ax.set_title("SG10K-SV r1.3 overlap Gnomad-SV")

ax.legend(handles=legend_elements, loc='upper right')
```
