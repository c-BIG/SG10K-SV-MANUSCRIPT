```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
df = pd.DataFrame({
    'SVTYPE': ["DEL", "DUP", "INS"],
    'SG10K-SV r1.3': [85195, 32464, 35670],
    'Gnomad-SV': [19274, 7479, 12584]
})
```

```python
## Colour code plan
## sg10ksv_palette = c(
#    "#D62628",    # SG-Chinese
#    "#9467BD",    # SG-Indian
#    "#2EA02B",    # SG-Malay
#    "#4DBBD5FF",  # DEL
#    "#DE8F44",    # INS
##    "#00A087FF"   # DUP
##)
gnomad_color = ['#4DBBD5FF','#00A087FF', '#DE8F44']

sg10k_color = ['#97FFFF','#C1FFC1', '#FFEC8B']

# Control legend color
# https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html
del_Gnomad_patch = mpatches.Patch(color= gnomad_color[0] )
dup_Gnomad_patch = mpatches.Patch(color= gnomad_color[1] )
ins_Gnomad_patch = mpatches.Patch(color= gnomad_color[2] )

del_SG10K_patch = mpatches.Patch(color= sg10k_color[0] )
dup_SG10K_patch = mpatches.Patch(color= sg10k_color[1] )
ins_SG10K_patch = mpatches.Patch(color= sg10k_color[2] )
```

```python
ax = plt.subplots()

# plotting columns
ax = sns.barplot(x=df["SVTYPE"], y=df["SG10K-SV r1.3"],  palette = sg10k_color , label="SG10K-SV r1.3" )
ax = sns.barplot(x=df["SVTYPE"], y=df["Gnomad-SV"],  palette = gnomad_color , label= "Gnomad-SV")
 
# renaming the axes
ax.set(xlabel="SVTYPE", ylabel="Count")
ax.set_title("SG10K-SV r1.3 overlap Gnomad-SV")


plt.legend(loc='upper right', \
           labels=['DEL (Gnomad)', 'DUP (Gnomad)', 'INS (Gnomad)', 'DEL (SG10K-SV)', 'DUP (SG10K-SV)', 'INS (SG10K-SV)'] , \
           handles= [del_Gnomad_patch, dup_Gnomad_patch, ins_Gnomad_patch, del_SG10K_patch, dup_SG10K_patch, ins_SG10K_patch])

# Jupyter notebook generate image file MUST be in same cell as plot
plt.savefig("overlap_gnomad.svg")
plt.savefig('overlap_gnomad.pdf')  
```
